### Compute trends for od and emi per species

#### read od by species

```python
coloc = '' #'' '_cs'
if coloc == '_cs':
    cs = True
else:
    cs = False
```

```python
fn0 = 'cache/od_species'+coloc+'.pkl'
if os.path.isfile(fn0):
    print('use',fn0)
    # for reading also binary mode is important 
    pklfile = open(fn0, 'rb')      
    MOD_OD = pickle.load(pklfile) 
    pklfile.close()
else:    
    MOD_OD = {}
    for mod_var in ['od550aer', 'od550dust', 'od550ss', 'od550so4', 'od550oa', 'od550bc']:
        MOD_OD[mod_var] = {}
        #fill paameters for this variable
        params = fill_params(params, mod_var)
        params['kind'] = 'mod' 
        params['min_dim'] = 0
        params['min_ntrend'] = 4

        for model in params['models']:
            mod_source = model

            #check if model in cache
            fn = 'cache/'+mod_source+'_'+mod_var+'.pkl'
            if os.path.isfile(fn):
                print(model,'use',fn)
                # for reading also binary mode is important 
                pklfile = open(fn, 'rb')      
                mod_data = pickle.load(pklfile) 
                pklfile.close()
            else:
                print(model,'read netcf file')
                reader = pya.io.ReadGridded(mod_source)
                mod_data = reader.read_var(mod_var, ts_type='daily')
                mod_data = mod_data.resample_time(to_ts_type='monthly')

                #write picke file in cache directory
                pklfile = open(fn, 'ab') 

                # source, destination 
                pickle.dump(mod_data, pklfile)                      
                pklfile.close()

            #crop the cube to interest period, so can handle WORLD region
            mod_data = mod_data.crop(time_range=(params['period'].split('-')[0], str(int(params['period'].split('-')[1])+1)))

            #space colocation
            _, MOD_OD[mod_var][model], _ = process_trend(
                mod_data, params, obs=obs_data,
                colocate_time=False, colocate_space=cs, 
                OBS_DF = OBS_DF,
                plot=False, show_plot=False, save_plot=False, write_json=False
            )
    
    #write picke file in cache directory
    pklfile = open(fn0, 'ab') 

    # source, destination 
    pickle.dump(MOD_OD, pklfile)                      
    pklfile.close()
```

#### read emi per species

```python
fn1 = 'cache/emi_species'+coloc+'.pkl'
if os.path.isfile(fn1):
    print('use',fn1)
    # for reading also binary mode is important 
    pklfile = open(fn1, 'rb')      
    MOD_EMI = pickle.load(pklfile) 
    pklfile.close()
else:    
    MOD_EMI = {}
    for mod_var in ['emidust', 'emiss', 'emiso4', 'emioa']:
        MOD_EMI[mod_var] = {}
        #fill paameters for this variable
        params = fill_params(params, mod_var)
        params['kind'] = 'mod' 
        params['min_dim'] = 0
        params['min_ntrend'] = 4

        for model in params['models']:
            mod_source = model

            #check if model in cache
            fn = 'cache/'+mod_source+'_'+mod_var+'.pkl'
            if os.path.isfile(fn):
                print(model,'use',fn)
                # for reading also binary mode is important 
                pklfile = open(fn, 'rb')      
                mod_data = pickle.load(pklfile) 
                pklfile.close()
            else:
                print(model,'read netcf file')
                reader = pya.io.ReadGridded(mod_source)
                mod_data = reader.read_var(mod_var, ts_type='daily')
                mod_data = mod_data.resample_time(to_ts_type='monthly')

                #write picke file in cache directory
                pklfile = open(fn, 'ab') 

                # source, destination 
                pickle.dump(mod_data, pklfile)                      
                pklfile.close()

            #crop the cube to interest period, so can handle WORLD region
            mod_data = mod_data.crop(time_range=(params['period'].split('-')[0], str(int(params['period'].split('-')[1])+1)))

            #space colocation
            _, MOD_EMI[mod_var][model], _ = process_trend(
                mod_data, params, obs=obs_data,
                colocate_time=False, colocate_space=cs, 
                OBS_DF = OBS_DF,
                plot=False, show_plot=False, save_plot=False, write_json=False
            )
    
    #write picke file in cache directory
    pklfile = open(fn1, 'ab') 

    # source, destination 
    pickle.dump(MOD_EMI, pklfile)                      
    pklfile.close()
```

### Plot the absolute trends

```python
#prepare dictionaries to be plotted
unit = ''
od_bars, od_errs, od_xpos, od_stats = {}, {}, {}, {}

barWidth = 0.10

#bars for od
var = 'od550oa'
#models = [model for model in list(MOD_OD[var].keys()) if model in MOD_OD[var] and '2000-2014' in MOD_OD[var][model]['WORLD']['trends']['trends']]
models = list(MOD_OD[var].keys())

od_vars = ['od550aer', 'od550dust', 'od550ss', 'od550so4', 'od550oa', 'od550bc']
for i, var in enumerate(od_vars):
    od_grps = [model for model in models if '2000-2014' in MOD_OD[var][model]['WORLD']['trends']['trends']]
    od_bars[var] = [MOD_OD[var][model]['WORLD']['trends']['trends']['2000-2014']['abs_slp']*10  for model in models if '2000-2014' in MOD_OD[var][model]['WORLD']['trends']['trends']]
    od_errs[var] = [0  for model in models if '2000-2014' in MOD_OD[var][model]['WORLD']['trends']['trends']]
    #compute the average and stdv
    od_stats[var] = '{:+4.1e} +/-{:4.1e}{} /dec'.format(np.nanmean(od_bars[var]),np.nanstd(od_bars[var]),' '+unit)
    if i==0:
        od_xpos[var] = np.arange(len(od_bars[var]))
    else:
        od_xpos[var] = [x + barWidth for x in od_xpos[od_vars[i-1]]]
```

```python
#bars for emi
#unit = 'kg m-2 s-1'
unit = 'kg.m-2.yr-1'
emi_bars, emi_errs, emi_xpos, emi_stats = {}, {}, {}, {}

emi_vars = ['emidust', 'emiss', 'emiso4', 'emioa', 'emibc']
for i, var in enumerate(emi_vars):
    print(var)
    emi_grps = models
    emi_bars[var] = [MOD_EMI[var][model]['WORLD']['trends']['trends']['2000-2014']['abs_slp']*60*60*24*365*10 if var in MOD_EMI and model in MOD_EMI[var] and '2000-2014' in MOD_EMI[var][model]['WORLD']['trends']['trends'] else np.nan for model in models ]
    emi_errs[var] = [0  for model in models]
    #compute the average and stdv
    emi_stats[var] = '{:+4.1e} +/-{:4.1e}{}/dec'.format(np.nanmean(emi_bars[var]),np.nanstd(emi_bars[var]),' '+unit)
    if i==0:
        emi_xpos[var] = np.arange(len(emi_bars[var]))
    else:
        emi_xpos[var] = [x + barWidth for x in emi_xpos[emi_vars[i-1]]]
        
#crop group names
od_grps = [grp.split('-')[0] for grp in od_grps]
od_grps = [grp.split('_')[0] for grp in od_grps]




#pick one color per species from palette
palette = sns.color_palette("deep", 18)

colors = {
    'od550aer': (0, 0, 0),
    'od550dust': palette[1],
    'od550ss': palette[0],
    'od550so4': palette[3],
    'od550oa': palette[2],
    'od550bc': palette[7],
    'emidust': palette[1],
    'emiss': palette[0],
    'emiso4': palette[3],
    'emioa': palette[2],
    'emibc': palette[7]
}
```

```python
fs = 0.8
nl, nc = 2, 1
fig, axs = plt.subplots(nl, nc, sharex='col', sharey=False, 
                       gridspec_kw={
                           'wspace': 0.01,
                           'hspace': 0.05
                       })


#add a big axes, hide frame
fig.add_subplot(111, frameon=False)
il, ic = 0, 0

for il, typ in enumerate(['OD', 'EMISSIONS']):
    print(il, typ)
    ax = axs[il]
    if typ=='OD':
        unit = ''
        xpos, bars, errs, stats, grps = od_xpos, od_bars, od_errs, od_stats, od_grps
        varss = od_vars
        ylabel = 'Trends ('+unit+'/dec)'
        if coloc=='_cs':
            ymin, ymax = -5e-2, 5e-2
        else:
            ymin, ymax = -5e-3, 5e-3
    elif typ=='EMISSIONS':
        unit = 'kg.m-2.yr-1'
        xpos, bars, errs, stats, grps = emi_xpos, emi_bars, emi_errs, emi_stats, od_grps
        varss = emi_vars
        ylabel = 'Trends ('+unit+'/dec)'
        if coloc=='_cs':
            ymin, ymax = -5e-5, 5e-5
        else:
            ymin, ymax = -5e-4, 5e-4

    #plot bars
    for var in varss:
        params = fill_params(params, var)
        if params['ylabel']=='AOD':
            lab = 'Total'+': '+stats[var]
        else:
            lab = params['ylabel'].split('_')[0]+': '+stats[var]
        ax.bar(xpos[var], bars[var], yerr=errs[var], color=(*colors[var],0.7), ecolor=colors[var], width=barWidth, edgecolor='white', label=lab)
        #if bar empty, plot an x
        for i, bar in enumerate(bars[var]):
            if np.isnan(bar):
                ax.plot(xpos[var][i],0,'x',color=colors[var])
    
    # Add xticks on the middle of the group bars
    ax.set_ylabel(ylabel)
    #if typ=='EMISSIONS':
    ax.set_xticks([r + barWidth for r in range(len(models))])
    ax.set_xticklabels([mod.split('_')[0] for mod in models],ha='right',rotation=45)
        

    ax.set_ylim(ymin, ymax)
    
    # Shrink current axis by 20%
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

    # Put a legend to the right of the current axis
    ax.legend(loc='center left', facecolor='#F1F1F1', bbox_to_anchor=(1, 0.5))


    #setbackground color
    ax.set_facecolor('#F1F1F1')

    #grid
    ax.grid(False)
    ax.grid(True, which='major', axis='both')

    ax.text(0.95,0.9,typ,ha='right', fontsize=18, fontweight='bold', transform=ax.transAxes)

#change size of the figure
# hide tick and tick label of the big axes
plt.tick_params(labelcolor=None, top=False, bottom=False, left=False, right=False)
plt.grid(False)
plt.xticks([])
plt.yticks([])


fig.set_size_inches(9,9)
fig = plt.gcf()

plt.savefig('figs/species_trends'+coloc+'.png', dpi=300, bbox_inches='tight')
```
